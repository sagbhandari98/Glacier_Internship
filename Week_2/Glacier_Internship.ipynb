## Project Description:
The objective is to assist a private firm in the US cab industry to identify the most suitable company for investment. This involves conducting market research and analyzing data to determine the optimal investment choice between two cab companies.

Data Analysis Steps:

1. **Data Set Investigation and Joining:** Explore the provided datasets and perform necessary data joins.

2. **Exploratory Data Analysis (EDA) and Visualization:** Conduct EDA to gain insights into the data and visualize key trends and patterns.

3. **Hypothesis Testing:** Perform hypothesis tests to make informed decisions based on statistical analysis.

## Datasets Provided:

1. **Cab_Data.csv:** Contains transaction details for two cab companies.
2. **Customer_ID.csv:** A mapping table linking customer demographic details to unique identifiers.
3. **Transaction_ID.csv:** A mapping table linking transactions to customers and payment modes.
4. **City.csv:** Includes a list of US cities along with their populations and the number of cab users.

## Part 2: Merge and Visualize the Whole Dataset

### 2.1: Merge the Whole Dataset

```python

# Perform join
merged_data = pd.merge(transaction_data, customer_data, on='Customer ID', how='inner')
merged_data = pd.merge(merged_data, cab_data, on='Transaction ID', how='inner')

# Handle missing values if any
merged_data.dropna(inplace=True)

# Validate results
print(merged_data.head())
```

```python
merged_data
```

```python
# Perform join
final_data = pd.merge(merged_data, city_data, on='City')

# Handle missing values if any
final_data.dropna(inplace=True)

# Validate results
print(final_data.head())
```

```python
final_data
```

```python
# Convert 'Date' attribute to datetime format
final_data['Date'] = pd.to_datetime(final_data['Date of Travel'])

# Extract year from 'Date' attribute and create a new attribute 'Year'
final_data['Year'] = final_data['Date'].dt.year

# Display the DataFrame with the new 'Year' attribute
print(final_data.head())
```

## Analysis of Transaction

```python
# Create a 2x2 grid of plots
fig = plt.figure(figsize=(15, 12))

# Total Transactions by Company
fig.add_subplot(2, 2, 1)
group = final_data.groupby('Company')['Transaction ID'].count().reset_index()
ax1 = sns.barplot(data=group, x='Company', y='Transaction ID', palette="magma")
ax1.set_title("Company-wise Transaction Volume", fontsize=14)
ax1.set_xlabel("Company", fontsize=12)
ax1.set_ylabel("Transactions Count", fontsize=12)
ax1.bar_label(ax1.containers[0])

# Total Transactions by Gender for each Company
fig.add_subplot(2, 2, 2)
group = final_data.groupby(by=['Company', 'Gender'])['Transaction ID'].count().reset_index()
ax2 = sns.barplot(data=group, x="Company", hue="Gender", y='Transaction ID', palette="magma")
ax2.set_title("Gender-wise Transaction Distribution", fontsize=14)
ax2.set_xlabel("Company", fontsize=12)
ax2.set_ylabel("Transactions Count", fontsize=12)
for container in ax2.containers:
    ax2.bar_label(container)

# Total Transactions by Payment Mode for each Company
fig.add_subplot(2, 2, 3)
group = final_data.groupby(by=['Company', 'Payment_Mode'])['Transaction ID'].count().reset_index()
ax3 = sns.barplot(data=group, x="Company", hue="Payment_Mode", y='Transaction ID', palette="magma")
ax3.set_title("Transactions by Payment Mode", fontsize=14)
ax3.set_xlabel("Company", fontsize=12)
ax3.set_ylabel("Transactions Count", fontsize=12)
for container in ax3.containers:
    ax3.bar_label(container)

# Total Profit by Company
fig.add_subplot(2, 2, 4)
group = final_data.groupby(by=['Company'])['Profit'].sum().reset_index()
ax4 = sns.barplot(data=group, x="Company", y='Profit', palette="magma")
ax4.set_title("Profit Distribution Across Companies", fontsize=14)
ax4.set_xlabel("Company", fontsize=12)
ax4.set_ylabel("Total Profit (USD)", fontsize=12)
ax4.bar_label(ax4.containers[0])

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

**Graph 1: Company-wise Transaction Volume, Gender-wise Transaction Distribution by Company, Transactions by Payment Mode for Each Company, Total Profit by Company**

## Analysis  City-Wise

```python
# Create a 2x1 grid of plots
fig, axs = plt.subplots(2, 1, figsize=(15, 15))

# Total Population by City
population = final_data.groupby("City").agg({"Population":"count"}).sort_values('Population').reset_index()
ax1 = sns.barplot(data=population, x='Population', y='City', palette="magma", ax=axs[0])
ax1.set_title("Total Population by City", fontsize=14)
ax1.set_xlabel("Total Population", fontsize=12)
ax1.set_ylabel("City", fontsize=12)

# Total Income by City
group_income = final_data.groupby('City')['Income (USD/Month)'].sum().reset_index()
ax2 = sns.barplot(data=group_income, x='Income (USD/Month)', y='City', palette="magma", ax=axs[1])
ax2.set_title("Total Income by City", fontsize=14)
ax2.set_xlabel("Total Income (USD)", fontsize=12)
ax2.set_ylabel("City", fontsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

**Graph 2: Total Population and Income by City**

```python
# Filter data for Yellow and Pink cab companies
filtered_data = final_data[final_data['Company'].isin(['Yellow Cab', 'Pink Cab'])]

# Group by City and Company to get the count of transactions
grouped_data = filtered_data.groupby(['City', 'Company'])['Transaction ID'].count().reset_index()

# Create a pivot table for easier plotting
pivot_data = grouped_data.pivot(index='City', columns='Company', values='Transaction ID').fillna(0)

# Plotting
plt.figure(figsize=(15, 10))
colors = sns.color_palette("magma", n_colors=2)
ax = pivot_data.plot(kind='bar', stacked=False, color=colors, figsize=(15, 10))

# Set plot title and labels
plt.title('Customer Presence of Yellow and Pink Cab Services City-wise', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Number of Transactions', fontsize=14)

# Add legend
plt.legend(title='Company', title_fontsize='13', fontsize='12')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

**Graph 3: Customer presence of Yellow and Pink cab services city-wise**

```python
plt.figure(figsize=(15, 10))

# Grouping the data
group = final_data.groupby(['Company', 'City'])['Profit'].sum().sort_values(ascending=False).reset_index()

# Plotting
ax = sns.barplot(data=group, x='City', y='Profit', hue='Company', palette='magma')

# Customizing the plot
plt.title("Profit in Cities by Company", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Total Profit (USD)", fontsize=14)
plt.xticks(rotation=45, ha='right')

# Adding legend with appropriate title
plt.legend(title='Company', title_fontsize='13', fontsize='12')

# Display the plot
plt.tight_layout()
plt.show()
```
